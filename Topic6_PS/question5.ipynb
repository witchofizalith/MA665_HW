$\textbf{5.}$  Consider the function $x(t) =  sin(2 pi t^2)$.That's a "t squared" input in the $sin$ function. 

$\textbf{5i.}$ Simulate the function $x(t) = \sin(2\pi t^2)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
```

```python
dt = 0.001  # Sampling interval
t = np.arange(0, 10, dt)
x = np.sin(2 * np.pi * t**2)

# Plot the function
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('x(t)')
plt.title('Function x(t) = sin(2πt^2)')
plt.grid(True)
plt.show()
```

$\textbf{5ii.}$ Compute the spectrum:

```python
frequencies = np.fft.fftfreq(len(t), dt)
X = np.fft.fft(x)

#absolute value to get the magnitude spectrum
magnitude_spectrum = np.abs(X)

plt.plot(frequencies, magnitude_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude Spectrum')
plt.title('Frequency Spectrum of x(t)')
plt.grid(True)
plt.show()
```

This kinda of resembles the spectrum graph made by EEG-4.mat. 

$\textbf{5iii.}$ Compute the spectrogram:

```python
# Compute the spectrogram
f, t_spec, Sxx = spectrogram(x, 1/dt)  # Use 1/dt as the sampling frequency

# Plot the spectrogram
plt.pcolormesh(t_spec, f, 10 * np.log10(Sxx))
plt.colorbar(label='Power/Frequency (dB/Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Spectrogram of x(t)')
plt.grid(True)
plt.show()
```

There appears to be consistent changes in activity levels across time for this function.
